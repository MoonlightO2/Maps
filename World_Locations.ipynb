```python
import pandas as pd
import numpy as np
import folium
from folium import Marker
import csv
from geopy.geocoders import GoogleV3
from geopy.geocoders import Nominatim
```

```python
file_path1 = '/content/drive/MyDrive/2. Projects/Colab Notebooks/Portfolio/All datasets for Data Analysis/All datasets/World Locations.xlsx'
file_path2 = '/content/drive/MyDrive/2. Projects/Colab Notebooks/Portfolio/All datasets for Data Analysis/All datasets/World Locations.csv'
map_path = '/content/drive/MyDrive/2. Projects/Colab Notebooks/Portfolio/All datasets for Data Analysis/All datasets/World Locations.html'
```

```python
data = pd.read_excel(file_path1)
print('Shape: ', data.shape, '\n')
print('Columns: ', data.columns.tolist(), '\n')
data.head()
```

```python
def get_coordinates(api_key, city, country):
    geolocator = GoogleV3(api_key=api_key)
    location = geolocator.geocode(f"{city}, {country}")

    if location:
        return location.latitude, location.longitude
    else:
        return None

# Replace 'your_api_key' with your actual Google Maps API key
api_key = 'AIzaSyD-sbQ-cCCXKE88286k6fxRfn0KVeQsjKk'

# Initialize an empty list to store locations
locations = []

# Iterate over rows in the DataFrame and extract relevant information
for index, row in data.iterrows():
    country = row['Country']
    city = row['City']

    # Get coordinates using the get_coordinates function
    coordinates = get_coordinates(api_key, city, country)

    # Append data to the locations list
    locations.append({
        'Country': country,
        'City': city,
        'Latitude': coordinates[0] if coordinates else 'Not found',
        'Longitude': coordinates[1] if coordinates else 'Not found',
    })

# Create a new DataFrame from the locations list
locations_df = pd.DataFrame(locations)

# Save the locations DataFrame to a new CSV file
locations_df.to_csv(file_path2, index=False)

print("Coordinates written to: ", file_path2)
```

```python
print('Shape: ', locations_df.shape, '\n')
print('Columns: ', locations_df.columns.tolist(), '\n')
locations_df.head()
```

```python
# Create a Folium Map centered at an average location
map_center = [locations_df['Latitude'].mean(), locations_df['Longitude'].mean()]
my_map = folium.Map(location=map_center, zoom_start=3)

# Add markers for each row in the DataFrame
for index, row in locations_df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row['Country']} - {row['City']}",
    ).add_to(my_map)

# Save the map to an HTML file or display it
my_map.save(map_path)
display(my_map)
```
